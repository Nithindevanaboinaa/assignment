```python
#1.7. Write a program to create a simple calculator performing only four basic operations(+,-,/,*).
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    return x / y

def calculator():
    print("Select operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    
    while True:
        choice = input("Enter choice (1/2/3/4): ")
        
        if choice in ['1', '2', '3', '4']:
            try:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
            except ValueError:
                print("Invalid input. Please enter numeric values.")
                continue

            if choice == '1':
                print(f"{num1} + {num2} = {add(num1, num2)}")
            elif choice == '2':
                print(f"{num1} - {num2} = {subtract(num1, num2)}")
            elif choice == '3':
                print(f"{num1} * {num2} = {multiply(num1, num2)}")
            elif choice == '4':
                print(f"{num1} / {num2} = {divide(num1, num2)}")
        else:
            print("Invalid choice. Please select a valid operation.")
        
        next_calculation = input("Do you want to perform another calculation? (yes/no): ").lower()
        if next_calculation != 'yes':
            break

calculator()
```

```python
#1.11. Write a program to find prime numbers between 2 to 100 using nested for loops.
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def find_primes(start, end):
    primes = []
    for num in range(start, end + 1):
        if is_prime(num):
            primes.append(num)
    return primes


start = 2
end = 100

prime_numbers = find_primes(start, end)
print(f"Prime numbers between {start} and {end} are: {prime_numbers}")
```

```python
"""
4. Create a simple banking system using object-oriented concepts in Python. Design classes representing different types of bank accounts such as savings and checking. Implement methods for deposit, withdraw, and balance inquiry. Utilize inheritance to manage different account types efficiently.

Tosk

1. Define base class(es) for bank accounts with common attributes and methods

2. Implement subclasses for specific account types (eg, Savings Account, CheckingAccount)

3. Provide methods for deposit, withdraw, and balance inquiry in each subciass

4. Test the banking system by creating instances of different account types and performing transactions.
"""

class BankAccount:
    def __init__(self, account_number, holder_name, balance=0):
        self._account_number = account_number
        self._holder_name = holder_name
        self._balance = balance  # Use a single underscore for protection

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited ${amount}. New balance: ${self._balance}")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money
    def withdraw(self, amount):
        if amount > 0:
            if amount <= self._balance:
                self._balance -= amount
                print(f"Withdrew ${amount}. New balance: ${self._balance}")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    # Method to check balance
    def check_balance(self):
        print(f"Account balance: ${self._balance}")

    # Getter for account number
    def get_account_number(self):
        return self._account_number

    # Getter for holder name
    def get_holder_name(self):
        return self._holder_name

class SavingsAccount(BankAccount):
    def __init__(self, account_number, holder_name, balance=0, interest_rate=0.01):
        super().__init__(account_number, holder_name, balance)
        self._interest_rate = interest_rate  # Use a single underscore for protection

    # Method to apply interest
    def apply_interest(self):
        interest = self._balance * self._interest_rate
        self.deposit(interest)
        print(f"Applied interest of ${interest}.")

class CheckingAccount(BankAccount):
    def __init__(self, account_number, holder_name, balance=0, overdraft_limit=500):
        super().__init__(account_number, holder_name, balance)
        self._overdraft_limit = overdraft_limit  # Use a single underscore for protection

    # Override withdraw method to include overdraft limit
    def withdraw(self, amount):
        if amount > 0:
            if (self._balance + self._overdraft_limit) >= amount:
                self._balance -= amount
                print(f"Withdrew ${amount}. New balance: ${self._balance}")
            else:
                print("Insufficient funds, including overdraft limit.")
        else:
            print("Withdrawal amount must be positive.")

# Create instances of different account types
savings_account = SavingsAccount("SA123", "Alice", 1000)
checking_account = CheckingAccount("CA123", "Bob", 500)

# Test SavingsAccount functionality
print("Savings Account:")
savings_account.check_balance()
savings_account.deposit(200)
savings_account.withdraw(100)
savings_account.apply_interest()
savings_account.check_balance()

# Test CheckingAccount functionality
print("\nChecking Account:")
checking_account.check_balance()
checking_account.deposit(150)
checking_account.withdraw(600)
checking_account.withdraw(100)
checking_account.check_balance()
```

```python
#27. In a partially destroyed laboratory record of an analysis of correlation data, the following results only are legible: 
#Variance of x = 9, Regression equations are: 
#    (i) 8x−10y = −66; 
#    (ii) 40x − 18y = 214. 
#    What are 
#   (a) the mean values of x and y, 
#    (b) the coefficient of correlation between x and y, 
#    (c) the σ of y.



import sympy as sp

def solve_regression_problem():
    x, y = sp.symbols('x y')
    eq1 = sp.Eq(8*x - 10*y, -66)
    eq2 = sp.Eq(40*x - 18*y, 214)

    # Solve the equations simultaneously for x and y
    mean_values = sp.solve((eq1, eq2), (x, y))
    mean_x = mean_values[x]
    mean_y = mean_values[y]

    # Variance of x
    variance_x = 9
    sigma_x = sp.sqrt(variance_x)

    # Calculate the slopes (regression coefficients)
    # From Equation 1: 8x - 10y = -66
    # 10y = 8x + 66
    # y = (8/10)x + 66/10
    b_yx = 8 / 10

    # From Equation 2: 40x - 18y = 214
    # 40x = 18y + 214
    # x = (18/40)y + 214/40
    b_xy = 18 / 40

    # Calculate the coefficient of correlation r
    r_squared = b_yx * b_xy
    r = sp.sqrt(r_squared)

    # Calculate the standard deviation of y (sigma_y)
    sigma_y = (b_yx * sigma_x) / r

    # Return the results
    return {
        'mean_x': mean_x,
        'mean_y': mean_y,
        'correlation_coefficient': r,
        'sigma_y': sigma_y
    }

# Get the results
results = solve_regression_problem()

# Print the results
print(f"Mean value of x: {results['mean_x']}")
print(f"Mean value of y: {results['mean_y']}")
print(f"Coefficient of correlation: {results['correlation_coefficient']}")
print(f"Standard deviation of y: {results['sigma_y']}")
```

```python
#31. The mean of a distribution is 60 with a standard deviation of 10. Assuming that the distribution is normal, 
#what percentage of items be (i) between 60 and 72, (ii) between 50 and 60, (iii) beyond 72 and (iv) between 
#70 and 80?
import scipy.stats as stats

# Given mean and standard deviation
mean = 60
std_dev = 10

# Function to calculate the percentage of items in a given range
def percentage_in_range(mean, std_dev, lower_bound, upper_bound):
    lower_cdf = stats.norm.cdf(lower_bound, mean, std_dev)
    upper_cdf = stats.norm.cdf(upper_bound, mean, std_dev)
    return (upper_cdf - lower_cdf) * 100

# (i) Percentage of items between 60 and 72
percentage_60_72 = percentage_in_range(mean, std_dev, 60, 72)

# (ii) Percentage of items between 50 and 60
percentage_50_60 = percentage_in_range(mean, std_dev, 50, 60)

# (iii) Percentage of items beyond 72
percentage_beyond_72 = (1 - stats.norm.cdf(72, mean, std_dev)) * 100

# (iv) Percentage of items between 70 and 80
percentage_70_80 = percentage_in_range(mean, std_dev, 70, 80)

# Print the results
print(f"Percentage of items between 60 and 72: {percentage_60_72:.2f}%")
print(f"Percentage of items between 50 and 60: {percentage_50_60:.2f}%")
print(f"Percentage of items beyond 72: {percentage_beyond_72:.2f}%")
print(f"Percentage of items between 70 and 80: {percentage_70_80:.2f}%")
```

```python
#33. If the height of 500 students are normally distributed with mean 65 inch and standard deviation 5 inch. 
#How many students have height : a) greater than 70 inch. b) between 60 and 70 inch.
from scipy.stats import norm
mean = 65
std_dev = 5
total_students = 500

# a) Probability of height greater than 70 inches
prob_greater_than_70 = 1 - norm.cdf(70, mean, std_dev)
students_greater_than_70 = prob_greater_than_70 * total_students

# b) Probability of height between 60 and 70 inches
prob_between_60_and_70 = norm.cdf(70, mean, std_dev) - norm.cdf(60, mean, std_dev)
students_between_60_and_70 = prob_between_60_and_70 * total_students

# Print results
print(f"Number of students with height greater than 70 inches: {students_greater_than_70:.0f}")
print(f"Number of students with height between 60 and 70 inches: {students_between_60_and_70:.0f}")
```

```python
#35.A random sample of size 25 from a population gives the sample standard derivation to be 9.0. Test the 
#hypothesis that the population standard derivation is 10.5.
import scipy.stats as stats
n = 25
s = 9.0
sigma_0 = 10.5
# Hypothesized population variance
sigma_0_sq = sigma_0 ** 2
# Sample variance
s_sq = s ** 2
# Chi-square test statistic
chi_square_statistic = (n - 1) * s_sq / sigma_0_sq
# Degrees of freedom
df = n - 1
# Critical values and p-value for a two-tailed test
alpha = 0.05  # significance level
# Compute critical values for a two-tailed test
chi2_critical_low = stats.chi2.ppf(alpha / 2, df)
chi2_critical_high = stats.chi2.ppf(1 - alpha / 2, df)
# Compute p-value
p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, df), 1 - stats.chi2.cdf(chi_square_statistic, df))
print(f"Chi-square Test Statistic: {chi_square_statistic:.2f}")
print(f"Critical Values: ({chi2_critical_low:.2f}, {chi2_critical_high:.2f})")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
#37.100 students of a PW IOI obtained the following grades in Data Science paper :Grade :[A, B, C, D, E] Total Frequency :[15, 17, 30, 22, 16, 100]Using the  χ 2 test , examine the hypothesis that the distribution of grades is uniform
import scipy.stats as stats
# Given data
grades = ['A', 'B', 'C', 'D', 'E']
observed_frequencies = [15, 17, 30, 22, 16]
total_students = 100
number_of_grades = len(grades)

# Expected frequency if the distribution is uniform
expected_frequency = total_students / number_of_grades

# Calculate the chi-square test statistic
chi_square_statistic = sum((obs - expected_frequency) ** 2 / expected_frequency for obs in observed_frequencies)

# Degrees of freedom
df = number_of_grades - 1

# Compute p-value
p_value = 1 - stats.chi2.cdf(chi_square_statistic, df)

# Print results
print(f"Chi-square Test Statistic: {chi_square_statistic:.2f}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis. The distribution of grades is not uniform.")
else:
    print("Fail to reject the null hypothesis. The distribution of grades is uniform.")
```

```python
"""42. How can you implement user authentication in a Flask application?"""
 
#1.Install Flask-Login: Use pip to install it: ""pip install flask-login""

#2.Create a Flask application: Write the following code in app.py.

from flask import Flask, render_template, redirect, url_for, request
from flask_login import LoginManager, UserMixin, login_user, login_required, logout_user, current_user

app = Flask(__name__)
app.secret_key = 'supersecretkey'
login_manager = LoginManager()
login_manager.init_app(app)

# Dummy user store
users = {'user': {'password': 'pass'}}

class User(UserMixin):
    def __init__(self, username):
        self.id = username

@login_manager.user_loader
def load_user(username):
    return User(username) if username in users else None

@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']
        if users.get(username, {}).get('password') == password:
            user = User(username)
            login_user(user)
            return redirect(url_for('profile'))
    return '''
        <form method="POST">
            Username: <input type="text" name="username"><br>
            Password: <input type="password" name="password"><br>
            <input type="submit" value="Login">
        </form>
    '''

@app.route('/profile')
@login_required
def profile():
    return f'Hello, {current_user.id}! <a href="/logout">Logout</a>'

@app.route('/logout')
@login_required
def logout():
    logout_user()
    return redirect(url_for('login'))

if __name__ == '__main__':
    app.run(debug=True)
```

```python
"""43. Describe the process of connecting a Flask app to a SQLite database using SQLAlchemy."""

#Install Flask-SQLAlchemy: Use pip to install it:  pip install flask-sqlalchemy

#Create a Flask application: Write the following code in app.py.

from flask import Flask, request
from flask_sqlalchemy import SQLAlchemy

app = Flask(__name__)
app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///example.db'
db = SQLAlchemy(app)

class User(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    username = db.Column(db.String(80), unique=True, nullable=False)

@app.route('/add_user', methods=['POST'])
def add_user():
    username = request.form['username']
    new_user = User(username=username)
    db.session.add(new_user)
    db.session.commit()
    return 'User added!'

if __name__ == '__main__':
    db.create_all()  # Creates database and tables
    app.run(debug=True)
```
