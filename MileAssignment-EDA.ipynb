```python
#Loading the Data
import pandas as pd
df = pd.read_csv('dataset.csv')
df.head()
```

```python
#Summary Statistics and Structure
df.describe(include='all')

df.info()
```

```python
#Missing values
df.isnull().sum()
```

```python
#Categorial Data Analysis

categorical_columns = ['GENDER', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE',
                       'CHRONIC_DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING',
                       'ALCOHOL_CONSUMING', 'COUGHING', 'SHORTNESS_OF_BREATH',
                       'SWALLOWING_DIFFICULTY', 'CHEST_PAIN', 'LUNG_CANCER']

for col in categorical_columns:
    print(df[col].value_counts())
    print()
```

```python
#Numerical data analysis

import matplotlib.pyplot as plt
import seaborn as sns

df['AGE'].hist(bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

numeric_df = df.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Relationship Analysis

for col in categorical_columns:
    if col != 'LUNG_CANCER':
        sns.countplot(x=col, hue='LUNG_CANCER', data=df)
        plt.title(f'Relationship between {col} and Lung Cancer')
        plt.show()

pd.crosstab(df['GENDER'], df['SMOKING'], normalize='index').plot(kind='bar', stacked=True)
plt.title('Interaction between Gender and Smoking')
plt.xlabel('Gender')
plt.ylabel('Proportion')
plt.show()
```

```python
#Pairplot

sns.pairplot(df, hue='LUNG_CANCER')
plt.show()
```

```python
#Target Variable Analysis
df['LUNG_CANCER'].value_counts().plot(kind='bar')
plt.title('Distribution of Lung Cancer')
plt.xlabel('Lung Cancer')
plt.ylabel('Count')
plt.show()
```
